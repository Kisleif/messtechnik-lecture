# Grundlagen

## Einleitung

Begriffserklärung: Messobjekt hat eine Messgröße (physikalische Größe, Temp, I, …) von der der Messwert bestimmt werden sollen -> wir benötigen ein Verfahren um die Größe zu extrahieren -> Messgerät
Beispiel Temperaturmessung: Messtechnik an bestimmten Ort herrschende „Messgröße“ mittels eines geeigneten Aufbaus in einen „Messwert“ überführen. 
Der Messwert kann uns direkt angezeigt werden oder er kann nachfolgenden elektronischen Systemen zur Weiterverarbeitung in geeigneter Form zugeführt werden. 
Messsignale: Direkt mit der Messgröße zusammenhängende elektrische Signale zwischen den beiden Stellen, an denen Messgröße und Messwert anfallen. 
Messsignale tragen die Information über die Messgröße, auf unterschiedlichste Weise realisiert z.B. als analoger Spanungs- oder Stromwert, als frequenzmoduliertes Signal, als Digitalwort etc.
Messwert wird bestimmt und entspricht im Rahmen der MT einem Vielfachen einer Einheit
Mit Einheiten beschäftigen wir uns gleich noch, zunächst:
Messgerät bestimmt Vielfaches einer Einheit, damit dies überall auf der Welt gleich gut funktioniert muss ein Messgerät entsprechend geeicht / kalibriert werden
Gibt auch den Begriff des Justierens. Dieser beschreibt die Anpassung eines Messgerät an verschieden Umgebungen. EVtl. Müssen Messgeräte bei unterschiedlichen Temperaturen anders behandelt werden
Beispiel eines Blockdiagramms für diese Temperaturmessung
Beispiel zur Heizungsanlage erweitern (Prozesstechnik)
Wie man an der hier dargestellten Temperaturmessung sieht, sind die Werte nicht unbedingt exakt. 
Befassen uns später noch mal genauer mit diesen Messabweichungen und Begrifflichkeit

## Wunder an Präzision (2600 v. Chr.)

Weitergehende Beispiele aus dem Bereich der Wunder der Messtechnik werden hier kurz dargestellt. 
Cheops-Pyramide: älteste und größte der 3 Pyramiden von Gizeh. Grabmal für den ägypt. König (Pharao) Cheops errichtet, der während der 4. Dynastie im Alten Reich regierte (2620-2580 v. Chr.), gehört zu den 7 Weltwundern der Antike. Eines der einzigen Weltwunder welches bis heute erhalten geblieben ist. 
Seitenlänge: 230,33m und Höhe: 146,59m (4000 Jahre lang höchstes Bauwerk der Welt)
Ihre Einmessung wurde in sehr hoher Genauigkeit vorgenommen, in nachfolgenden Bauten nicht mehr erreicht.
Sie ist genau nach den vier Himmelsrichtungen ausgerichtet, und der Unterschied in den Längen ihrer vier Seiten beträgt weniger als ein Promille.
Es sind ‚nur‘ drei Parameter maßgeblich, um die Präzision einer Pyramide zu bestimmen:
die waagrechte Ausrichtung des Fundaments und aller folgenden Bauschichten,
die Orientierung nach den Himmelsrichtungen,
die Seitenneigung der Flächen.
Alle drei Parameter müssen nicht nur bei Baubeginn exakt festgelegt, sondern vor allem auch während des Baus kontinuierlich kontrolliert und nachgemessen werden, sonst wird das ganze Bauwerk sichtbar unregelmäßig. Es genügt nicht, beispielsweise die horizontale Ausrichtung lediglich bei Baubeginn festzulegen und dann drauflos zu bauen. Bei der enormen Höhe der großen Pyramiden von Gizah würde ein sich wiederholender Messfehler von wenigen Millimetern nach oben hin multipliziert.
Erwägt man also mögliche Messmethoden im Hinblick auf ihre Tauglichkeit beim Pyramidenbau, müssen drei wesentliche Bedingungen erfüllt werden:
Die Messtechnik ist mit steinzeitlichem Werkzeug und Wissen möglich.
Sie ist realistisch geeignet, bei den beträchtlichen Dimensionen der Pyramiden zumindest die erreichte Genauigkeit zu liefern.
Sie ist auch in großer Höhe und auf kleiner Fläche anwendbar, so dass sie bei jeder neuen Schicht wiederholt werden kann.
